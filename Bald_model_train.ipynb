```python
import os
import cv2
import numpy as np
import json
import signal

# Класс исключения для таймаута
class TimeoutException(Exception):
    pass

# Обработчик сигналов таймаута
def timeout_handler(signum, frame):
    raise TimeoutException("Operation timed out")

def check_image(path, timeout=5):
    """
    Проверяет корректность изображения по пути, используя таймаут.

    Если файл существует, считывает его как бинарные данные, преобразует в numpy-массив
    и пытается декодировать через cv2.imdecode. Если операция не укладывается в timeout секунд,
    выбрасывается TimeoutException.

    Возвращает (True, "") если успешно, иначе (False, причина ошибки)
    """
    if not os.path.exists(path):
        return False, "Путь не существует"
    try:
        # Читаем файл как бинарные данные
        with open(path, 'rb') as f:
            file_bytes = f.read()
        img_array = np.frombuffer(file_bytes, np.uint8)

        # Устанавливаем таймаут
        signal.signal(signal.SIGALRM, timeout_handler)
        signal.alarm(timeout)  # Таймаут в секундах
        try:
            img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
        finally:
            signal.alarm(0)  # Сбрасываем таймаут
        if img is None:
            return False, "cv2.imdecode вернул None"
        return True, ""
    except TimeoutException:
        return False, "Timeout during imdecode"
    except Exception as e:
        return False, f"Ошибка: {e}"

# Основной блок проверки изображений
dataset_path = "/content/dataset.jsonl"

if not os.path.exists(dataset_path):
    print(f"[ERROR] Файл датасета не найден: {dataset_path}")
else:
    with open(dataset_path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    total_pairs = len(lines)
    valid_pairs = 0
    invalid_entries = []

    for i, line in enumerate(lines):
        try:
            item = json.loads(line)
            source_path = item.get("source", "").strip()
            target_path = item.get("target", "").strip()

            valid_source, reason_source = check_image(source_path)
            valid_target, reason_target = check_image(target_path)

            if valid_source and valid_target:
                valid_pairs += 1
            else:
                invalid_entries.append({
                    "pair_index": i,
                    "source": source_path,
                    "source_issue": reason_source,
                    "target": target_path,
                    "target_issue": reason_target
                })
        except json.JSONDecodeError as e:
            invalid_entries.append({
                "pair_index": i,
                "error": f"Ошибка декодирования JSON: {e}"
            })

    print(f"Всего пар: {total_pairs}")
    print(f"Валидных пар: {valid_pairs}")
    print(f"Невалидных пар: {len(invalid_entries)}")
    if invalid_entries:
        print("Детали невалидных пар:")
        for entry in invalid_entries:
            print(entry)
```
